```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_excel('5104tables.xlsx', sheet_name="Figure2")
df.head()
df.columns = df.columns.str.strip().str.lower()
```

```python
df_mse_melted = df.melt(id_vars='model', var_name='Dataset', value_name='MSE')
# df_mse_melted['Dataset'] = df_melted['Dataset'].str.title()
```

```python
df_mae_melted = df.melt(id_vars='model', var_name='Dataset', value_name='MAE')
# df_mae_melted['Dataset'] = df_melted['Dataset'].str.title()
```

```python
sns.set(style="white")
palette = sns.color_palette("Blues", n_colors=len(df_mse_melted['Dataset'].unique()))

fig, axes = plt.subplots(1, 2, figsize=(16, 6), sharey=True)

sns.barplot(
    data=df_mse_melted, 
    x='model', 
    y='MSE', 
    hue='Dataset', 
    palette=palette,
    ax=axes[0]
)
axes[0].set_title('Model Performance (MSE)', fontsize=14)
axes[0].set_xlabel('Model')
axes[0].set_ylabel('Error Value')

sns.barplot(
    data=df_mae_melted, 
    x='model', 
    y='MAE', 
    hue='Dataset', 
    palette=palette,
    ax=axes[1]
)
axes[1].set_title('Model Performance (MAE)', fontsize=14)
axes[1].set_xlabel('Model')
axes[1].set_ylabel('')

axes[1].legend(title='Dataset', loc='upper right', fontsize=10, title_fontsize=11)

plt.tight_layout()
plt.show()
```
